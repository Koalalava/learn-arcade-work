```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px ### for plotting the data on world map


path = 'dead.csv'
df = pd.read_csv(path)
df.info()
df.head(100)
```

```python
df.drop(['Series Name','Series Code','Country Code'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
world = df.groupby("Country Name")["2004","2005","2006","2007","2008"].sum().reset_index()
world.head()
```

```python
from matplotlib import container
top_5 = world.sort_values(by="2008", ascending=False).head(5)
top_5.head
```

```python
plt.figure(figsize=(12,10))
plot = sns.barplot(x=top_5['2008'], y=top_5['Country Name'])
plot.bar_label(plot.containers[0])

plt.show()


```
